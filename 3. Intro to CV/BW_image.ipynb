```python
import cv2
from google.colab.patches import cv2_imshow
# read image and display
img = cv2.imread('/content/gdrive/MyDrive/tzuyu1.jpg')
cv2_imshow(img)
```

```python
# turn image to grayscale and display
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
# write image to gdrive
cv2.imwrite('/content/gdrive/MyDrive/tzuyu1_gray.png', img_gray)
```
